```python
import matplotlib.pyplot as plt

def make_graph(ticker_symbol):
    stock_data = yf.Ticker(ticker_symbol)
    data = stock_data.history(period="max")

    plt.figure(figsize=(10,6))
    plt.plot(data.index, data['Close'], label='Tesla Stock Price')
    plt.title('Tesla Stock Price Over Time')  # Title of the graph
    plt.xlabel('Date')  # X-axis label
    plt.ylabel('Close Price (USD)')  # Y-axis label
    plt.legend()
    plt.grid(True)
    plt.show()

make_graph("TSLA")
```

```python
def make_graph(ticker_symbol):
    stock_data = yf.Ticker(ticker_symbol)

    stock_price = stock_data.history(period="max")

    financials = stock_data.financials

    revenue = financials.loc['Total Revenue']

    # Create a figure for two subplots
    plt.figure(figsize=(14, 6))
    
    # Subplot for Stock Price
    plt.subplot(1, 2, 1)
    plt.plot(stock_price.index, stock_price['Close'], label='GameStop Stock Price', color='blue')
    plt.title('GameStop Stock Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Close Price (USD)')
    plt.legend()
    plt.grid(True)
    
    # Subplot for Revenue
    plt.subplot(1, 2, 2)
    plt.plot(revenue.index, revenue.values, label='GameStop Revenue', color='green')
    plt.title('GameStop Revenue Over Time')
    plt.xlabel('Date')
    plt.ylabel('Revenue (USD)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
make_graph("GME")
```
